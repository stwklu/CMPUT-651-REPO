```
# Load experiment results
with open('/content/drive/My Drive/Colab Notebooks/651/og_train_err.pkl', 'rb') as f:
  og_err = pickle.load(f)
with open('/content/drive/My Drive/Colab Notebooks/651/og_train_ppl.pkl', 'rb') as f:
  og_ppl = pickle.load(f)
with open('/content/drive/My Drive/Colab Notebooks/651/dual_train_err.pkl', 'rb') as f:
  dual_err = pickle.load(f)
with open('/content/drive/My Drive/Colab Notebooks/651/dual_train_ppl_1.pkl', 'rb') as f:
  og_ppl_1 = pickle.load(f)
with open('/content/drive/My Drive/Colab Notebooks/651/dual_train_ppl_2.pkl', 'rb') as f:
  og_ppl_2 = pickle.load(f)
```

```
# Plot loss
%matplotlib inline
f = plt.figure(figsize=(16,8))
ax = f.add_subplot(1,2,1)
ax.plot(og_err,  label='OG')
ax.plot(dual_err,  label='DUAL')
ax.set_yscale('log')
ax.set_title('MSE.')
ax.legend()

ax = f.add_subplot(1,2,2)
ax.plot(og_ppl, label='OG')
ax.plot(og_ppl_1, label='DUAL 1')
ax.plot(og_ppl_2, label='DUAL 2')
ax.set_title('Perplexity.')
ax.legend()
```
